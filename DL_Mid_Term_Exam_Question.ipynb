# Question 02: [ Marks 20 ]

Manual Gradient Calculation

Given:

z = w*x + b <br>
y_pred = sigmoid(z) <br>
L = −[ y log(y_pred) + (1−y) log(1−y_pred) ] <br>

a) Write dL/dy_pred <br>
b) Write dy_pred/dz <br>
c) Show that dL/dz = y_pred − y <br>


**Given:**

$$z = w \cdot x + b$$

$$y_{pred} = \sigma(z) = \frac{1}{1 + e^{-z}}$$

$$L = -\left[ y \log(y_{pred}) + (1 - y) \log(1 - y_{pred}) \right]$$

---

### a) Write dL/dy_pred

The loss function $L$ with respect to the predicted value $y_{pred}$ is:

$$L = -\left[ y \log(y_{pred}) + (1 - y) \log(1 - y_{pred}) \right]$$

Differentiating $L$ with respect to $y_{pred}$ yields:

$$\frac{dL}{dy_{pred}} = -\left[ \frac{y}{y_{pred}} + \frac{1-y}{1 - y_{pred}} \cdot (-1) \right] = -\left[ \frac{y}{y_{pred}} - \frac{1-y}{1 - y_{pred}} \right]$$

$$\frac{dL}{dy_{pred}} = \frac{1-y}{1 - y_{pred}} - \frac{y}{y_{pred}}$$

---

### b) Write dy_pred/dz

The predicted value $y_{pred}$ is the sigmoid of $z$:

$$y_{pred} = \sigma(z) = \frac{1}{1 + e^{-z}}$$

The derivative of the sigmoid function with respect to $z$ is:

$$\frac{dy_{pred}}{dz} = y_{pred}(1 - y_{pred})$$

---

### c) Show that dL/dz = y_pred − y

To find $\frac{dL}{dz}$, we apply the chain rule using the results from parts (a) and (b):

$$\frac{dL}{dz} = \frac{dL}{dy_{pred}} \times \frac{dy_{pred}}{dz}$$

Substitute the expressions for $\frac{dL}{dy_{pred}}$ and $\frac{dy_{pred}}{dz}$:

$$\frac{dL}{dz} = \left( \frac{1-y}{1 - y_{pred}} - \frac{y}{y_{pred}} \right) \times \left( y_{pred}(1 - y_{pred}) \right)$$

Distribute the term $y_{pred}(1 - y_{pred})$ across the terms in the parenthesis:

$$\frac{dL}{dz} = \left( \frac{1-y}{1 - y_{pred}} \times y_{pred}(1 - y_{pred}) \right) - \left( \frac{y}{y_{pred}} \times y_{pred}(1 - y_{pred}) \right)$$

Simplify the terms:

$$\frac{dL}{dz} = (1 - y)y_{pred} - y(1 - y_{pred})$$

$$\frac{dL}{dz} = y_{pred} - yy_{pred} - y + yy_{pred}$$

The $yy_{pred}$ terms cancel out:

$$\boxed{\frac{dL}{dz} = y_{pred} - y}$$

# Question 03: [ Marks 15 ]

x = 3
y1 = x^2
y2 = 4*x
L = y1 + y2

a) Draw the computation graph. <br>
b) Compute dL/dx manually. <br>
c) Explain how would you do it using Autograd <br>

**Given:**

$$x = 3, \quad y_1 = x^2, \quad y_2 = 4x, \quad L = y_1 + y_2$$

---

### a) Computation Graph
```
        x
       / \
      /   \
  square   4x
    |        |
    y1       y2
      \     /
       \   /
         +
         |
         L
```

**Forward values:**

$$y_1 = 3^2 = 9$$

$$y_2 = 4 \times 3 = 12$$

$$L = 9 + 12 = 21$$

---

### b) Compute dL/dx Manually

$$L = x^2 + 4x$$

Using the Chain Rule:

$$\frac{dL}{dx} = \frac{dL}{dy_1}\frac{dy_1}{dx} + \frac{dL}{dy_2}\frac{dy_2}{dx}$$

Since $L = y_1 + y_2$:

$$\frac{dL}{dy_1} = 1, \quad \frac{dL}{dy_2} = 1$$

And:

$$\frac{dy_1}{dx} = 2x, \quad \frac{dy_2}{dx} = 4$$

Therefore:

$$\frac{dL}{dx} = 1(2x) + 1(4) = 2x + 4$$

Substituting $x = 3$:

$$\frac{dL}{dx} = 2(3) + 4 = \boxed{10}$$

---

### c) How to do it using Autograd

Autograd automatically computes the gradient by doing a backward pass through the computation graph.

**Steps:**
1. Create tensor with `requires_grad=True`
2. Do the forward computation
3. Call `L.backward()`
4. Check `x.grad`



````python
import torch

x = torch.tensor(3.0, requires_grad=True)

y1 = x**2
y2 = 4*x
L = y1 + y2

L.backward()

print(x.grad)
````

# Question 04: [ Marks 15 ]

Create a scalar tensor x = 3.0 with gradient tracking enabled. <br>
Compute y = x² + 4x. <br>
Perform backpropagation. <br>
Print the gradient of x. <br>

Write the full PyTorch code yourself. <br>

## Write Answer 04:

````python
import torch

# Step 1: Create scalar tensor with gradient tracking
x = torch.tensor(3.0, requires_grad=True)

# Step 2: Forward computation
y = x**2 + 4*x

# Step 3: Backpropagation
y.backward()

# Step 4: Print gradient
print("Value of y:", y.item())
print("Gradient of x (dy/dx):", x.grad.item())
````

Create two tensors:

b = 1.0 (requires_grad=True) <br>

Let input x = 3.0. <br>
Compute z = w*x + b. <br>
Use Sigmoid to get y_pred.
 <br>

Define Binary Cross Entropy manually (without built-in loss). <br>
Compute gradients using backward(). <br>
Print gradients of w and b. <br>

Write full code.

## Write Answer 05:

````python
import torch

# Step 1: Create parameters with gradient tracking
w = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(1.0, requires_grad=True)

# Step 2: Input and true label
x = torch.tensor(3.0)
y = torch.tensor(1.0)

# Step 3: Linear computation
z = w * x + b

# Step 4: Sigmoid
y_pred = torch.sigmoid(z)

# Step 5: Manual Binary Cross Entropy
loss = -( y * torch.log(y_pred) +
          (1 - y) * torch.log(1 - y_pred) )

# Step 6: Backpropagation
loss.backward()

# Step 7: Print results
print("z:", z.item())
print("y_pred:", y_pred.item())
print("Loss:", loss.item())
print("Gradient of w:", w.grad.item())
print("Gradient of b:", b.grad.item())
````

# Question 07: [ Marks 20 ]


Create tensor w = 2.0 (requires_grad=True). <br>

Compute two different losses sequentially and call backward() twice. <br>
Print w.grad. <br>

Do not clear gradients. <br>
Explain what you observe. <br>

Write full code.


## Write Answer 07:



````python
import torch

# Create tensor
w = torch.tensor(2.0, requires_grad=True)

# First loss
L1 = w**2
L1.backward()

print("After first backward:")
print("w.grad =", w.grad.item())

# Second loss
L2 = 3*w
L2.backward()

print("\nAfter second backward (without clearing gradients):")
print("w.grad =", w.grad.item())
````

In this experiment, I created a tensor w = 2.0 with requires_grad=True. First, I computed the loss L1 = w² and called backward(). Since the derivative of w² is 2w, at w = 2 the gradient becomes 4. So after the first backward pass, w.grad was 4.0.

Then I defined another loss L2 = 3w and again called backward() without clearing the previous gradients. The derivative of 3w is 3. However, instead of replacing the old gradient, PyTorch added this new gradient to the previous one. So the total gradient became 4 + 3 = 7.

From this, I observed that PyTorch accumulates gradients by default. It does not automatically reset them after calling backward(). That is why in training loops we usually use optimizer.zero_grad() before calling backward(), otherwise gradients will keep accumulating and may cause incorrect updates.
